```python
#mushroom data set decision tree
import pandas as pd
import numpy as np
user = "bfo"
mushroom_dataframe = pd.read_csv(f"C:/Users/{user}/Downloads/mushrooms.csv", delimiter = ';')
mushroom_dataframe
mushroom_dataframe.head()
```

```python
mushroom_dataframe.columns
```

```python
mushroom_dataframe = mushroom_dataframe.rename(columns={'class':"class", 'cap-diameter':"cap_diameter", 'cap-shape':"cap_shape", 'cap-surface':"cap_surface", 'capcolor':"cap_color", 
'does-bruise-or-bleed':"does_bruise_or_bleed", 'gill-attachment':"gill_attachment", 'gill-spacing':"gill_spacing", 'gill-color':"gill_color", 'stemheight':"stem_height", 
'stem-width':"stem_width", 'stem-root':"stem_root", 'stem-surface':"stem_surface", 'stem-color':"stem_color", 'veil-type':"veil_type", 'veilcolor':"veil_color", 
'has-ring':"has_ring", 'ring-type':"ring_type", 'spore-print-color':"Spore_print_color", 'habitat':"habitat", 'season':"season"})
```

```python
import sklearn as sl
from sklearn.model_selection import train_test_split
x = mushroom_dataframe.drop("class", axis=1)
x
```

```python
x.cap_shape=mushroom_dataframe.cap_shape.astype('category').cat.codes
x.cap_surface=mushroom_dataframe.cap_surface.astype('category').cat.codes
x.cap_color=mushroom_dataframe.cap_color.astype('category').cat.codes
x.gill_attachment=mushroom_dataframe.gill_attachment.astype('category').cat.codes
x.does_bruise_or_bleed=mushroom_dataframe.does_bruise_or_bleed.astype('category').cat.codes
x.gill_spacing=mushroom_dataframe.gill_spacing.astype('category').cat.codes
x.gill_color=mushroom_dataframe.gill_color.astype('category').cat.codes
x.stem_root=mushroom_dataframe.stem_root.astype('category').cat.codes
x.stem_surface=mushroom_dataframe.stem_surface.astype('category').cat.codes
x.stem_color=mushroom_dataframe.stem_color.astype('category').cat.codes
x.veil_type=mushroom_dataframe.veil_type.astype('category').cat.codes
x.veil_color=mushroom_dataframe.veil_color.astype('category').cat.codes
x.has_ring=mushroom_dataframe.has_ring.astype('category').cat.codes
x.ring_type=mushroom_dataframe.ring_type.astype('category').cat.codes
x.spore_print_color=mushroom_dataframe.spore_print_color.astype('category').cat.codes
x.habitat=mushroom_dataframe.habitat.astype('category').cat.codes
x.season=mushroom_dataframe.season.astype('category').cat.codes
x
```

```python
df_label = mushroom_dataframe["class"].copy()
y=np.ravel(df_label)
y
```

```python
#check if dataset label are balanced
df_label.value_counts()
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.5, random_state=50)
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)
tree.plot_tree(clf)
```

```python
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("mushrooms")
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=x.columns.tolist(),class_names=y,filled=True, rounded=True,special_characters=True)
graph = graphviz.Source(dot_data)
graph
```
