```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
wineDataset = datasets.load_wine(as_frame=True)
print(wineDataset.data)
print(wineDataset.target)
```

```python
wineDataset.data.shape
```

```python
i, j = np.where(wineDataset.data == wineDataset.data['proline'].max())
print(i, j)
print(wineDataset.data["proline"])
print(f"{wineDataset.data.max()}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(wineDataset.data, wineDataset.target, test_size=0.3,random_state=250) # 70% training and 30% test
gnb = GaussianNB()
#Train the model using the training sets
gnb.fit(X_train, y_train)
#Predict the response for test dataset
y_pred = gnb.predict(X_test)
print(y_test, "\n", y_pred)
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7.5))
ax1.plot(y_test, y_pred, **{"marker" : "x"})
ax1.set_title("100% accuracy")
ax2.plot( wineDataset.target, wineDataset.data, **{"marker" : "x"})
```

```python
X_train, X_test, y_train, y_test = train_test_split(wineDataset.data, wineDataset.target, test_size=0.3,random_state=110) # 70% training and 30% test
#Train the model using the training sets
gnb.fit(X_train, y_train)
#Predict the response for test dataset
y_pred = gnb.predict(X_test)
print(y_test, "\n", y_pred)
# Model Accuracy, how often is the classifier correct?
accuracy_score = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy_score)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7.5))
ax1.plot(y_test, y_pred, **{"marker" : "x"})
ax1.set_title(f"{accuracy_score}% accuracy")
```
