```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 4,cluster_std = 1.5, random_state = 4)
print(X)
print(y)
```

```python
plt.style.use("seaborn")
plt.figure(figsize = (10,10))
plt.scatter(X[:,0], X[:,1], c = y, marker= '*', s=100, edgecolors='black')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
knns = []
preds = []
amount = 100

for i in range(1, amount):
    knns.append(KNeighborsClassifier(n_neighbors=i))
    knns[i - 1].fit(X_train, y_train)
    preds.append(knns[i - 1].predict(X_test))
print(knns)
print(preds)
```

```python
for i in range(1, amount):
    print(f"Accuracy with k={i}", accuracy_score(y_test, preds[i - 1])*100)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 10))
for i in range():
    for j in range(0, int(amount/4)):
        axs[i, j].scatter(X_test[ :,0], X_test[:,1], c = preds[i*j], marker = "*", s = 100, edgecolors="black")
        axs[i, j].set_title(f"Predicted values with k={i*j}")
axs[0, 0].scatter(X_test[ :,0], X_test[:,1], c = preds[i*j], marker = "*", s = 100, edgecolors="black")
axs[0, 0].set_title(f"Predicted values with k={i*j}")
```
