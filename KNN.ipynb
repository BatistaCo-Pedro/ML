```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split 
```

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 4,cluster_std = 1.5, random_state = 4)
print(X)
print(y)
```

```python
plt.style.use("seaborn")
plt.figure(figsize = (10,10))
plt.scatter(X[:,0], X[:,1], c = y, marker= '*', s=100, edgecolors='black')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
knn5 = KNeighborsClassifier(n_neighbors = 5)
knn1 = KNeighborsClassifier(n_neighbors = 1)
knn10 = KNeighborsClassifier(n_neighbors = 10)
knn7 = KNeighborsClassifier(n_neighbors=7)

knn5.fit(X_train, y_train)
knn1.fit(X_train, y_train)
knn10.fit(X_train, y_train)
knn7.fit(X_train, y_train)
y_pred_5 = knn5.predict(X_test)
y_pred_1 = knn1.predict(X_test) 
y_pred_10 = knn10.predict(X_test)
y_pred_7 = knn7.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy with k=5", accuracy_score(y_test, y_pred_5)*100)
print("Accuracy with k=1", accuracy_score(y_test, y_pred_1)*100)
print("Accuracy with k=10", accuracy_score(y_test, y_pred_10)*100)
print("Accuracy with k=7", accuracy_score(y_test, y_pred_7)*100)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
axs[0, 0].scatter(X_test[ :,0], X_test[:,1], c = y_pred_1, marker = "*", s = 100, edgecolors="black")
axs[0,0].set_title("Predicted values with k=1")
axs[0, 1].scatter(X_test[:,0], X_test[:,1], c=y_pred_5, marker= '*', s=100,edgecolors='black')
axs[0, 1].set_title("Predicted values with k=5")
axs[1, 0].scatter(X_test[:,0], X_test[:,1], c=y_pred_10, marker= '*', s=100,edgecolors='black')
axs[1, 0].set_title("Predicted values with k=10")
axs[1, 1].scatter(X_test[:,0], X_test[:,1], c=y_pred_7, marker= '*', s=100,edgecolors='black')
axs[1, 1].set_title("Predicted values with k=7")
```
